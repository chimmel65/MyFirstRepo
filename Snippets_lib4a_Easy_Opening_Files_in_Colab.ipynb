**Dateien aus dem lokalen Dateisystem hochladen**<br>
files.upload gibt ein Wörterbuch der hochgeladenen Dateien zurück. Der Explorer öffnet sich und es kann eine Datei ausgewählt werden.

```python
# Dateien aus dem lokalen Dateisystem öffnen / hochladen
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

**Datei mit einer Adresse (URL) im WWW öffnen**
